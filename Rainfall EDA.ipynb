## UP rainfall analysis

```python
rain_UP = rainfall[rainfall['STATE/UT']=='UTTAR PRADESH']
rain_UP
```

```python
rain_UP.sort_values('ANNUAL', axis = 0, ascending = False)[0:10]
```

```python
plt.figure(figsize = (20,10))
plt.plot('DISTRICT','JAN+FEB', data = rain_UP, marker = 'o', label = 'Jan & Feb')
plt.plot('DISTRICT', 'MAM',data = rain_UP, marker = 'o', label = 'March , April & May')
plt.plot('DISTRICT', 'JJAS', data = rain_UP, marker = 'o', label = 'June, July, Aug & Sept')
plt.plot('DISTRICT', 'OND',data = rain_UP, marker = 'o', label = 'Oct, Nov & Dec')
plt.xticks(rotation = 90)
plt.legend()
plt.show()
```

```python
rain_UP2 = rain_UP.sort_values('ANNUAL', axis = 0 , ascending = True)
rain_UP2
```

```python
plt.figure(figsize = (20,5))
p = sns.barplot(x = 'DISTRICT', y= 'ANNUAL', data = rain_UP2)
p.tick_params(axis = 'x', labelrotation =90)
plt.xlabel('Districts in UP')
plt.ylabel('RAINFALLL IN MM')
plt.title('Rainfall in EAST UP & WEST UP')
plt.show()
```
